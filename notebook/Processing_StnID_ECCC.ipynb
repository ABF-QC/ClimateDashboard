<h4> Import necessary module

````python
import pandas as pd

from ipyleaflet import Map, basemaps, basemap_to_tiles, LayersControl, AwesomeIcon, Marker, CircleMarker, Popup, LegendControl

from ipywidgets import HTML, Layout
````

<h4> Reading Station ID Inventatory file

````python
# First 3 lines of the csv files must be discarded
rows2skip = list(range(0,3))

stnID_df = pd.read_csv('../data/ECCC/Station_Inventory_EN.csv', skiprows=rows2skip)
stnID_df
````

<h4> Keeping the station for the Quebec Province only

````python
stnID_df =stnID_df[stnID_df['Province'] == 'Quebec'.upper()]
stnID_df
````

````python
stn_reports = {
    'hly_dly_mly' : {
        'name' : 'HLY/DLY/MLY',
        'color' : 'purple',
    },
    'hly_dly' : {
        'name' : 'HLY/DLY',
        'color' : 'blue',
    },
    'hly_mly' : {
        'name' : 'HLY/MLY',
        'color' : 'black',
    },
    'hly' : {
        'name' : 'HLY',
        'color' : 'red',
    },
    'dly_mly' : {
        'name' : 'DLY/MLY',
        'color' : 'orange',
    },
    'mly' : {
        'name' : 'MLY',
        'color' : 'yellow',
    },
    'other' : {
        'name' : 'no reports',
        'color' : 'pink',
    },
}

stn_reports['hly_dly_mly']['name']
stn_reports['hly_dly_mly']['color']
````

````python
message = []
markers = []
for stn in stnID_df['Climate ID'] :
    lat = stnID_df[stnID_df['Climate ID'] == stn]['Latitude (Decimal Degrees)'].values[0]
    lon = stnID_df[stnID_df['Climate ID'] == stn]['Longitude (Decimal Degrees)'].values[0]
    
    #print(lat,lon)
    #markers.append(CircleMarker(location = (lat, lon), radius = 1, color = 'red', fill_color = 'red'))
    
    # Find availability of data
    hourly = stnID_df[stnID_df['Climate ID'] == stn]['HLY First Year'].notnull().values[0] 
    daily  = stnID_df[stnID_df['Climate ID'] == stn]['DLY First Year'].notnull().values[0] 
    monthly = stnID_df[stnID_df['Climate ID'] == stn]['MLY First Year'].notnull().values[0] 
    
    if hourly and daily and monthly :
        markers.append(CircleMarker(location = (lat, lon), radius = 1, 
                                    color = stn_reports['hly_dly_mly']['color'],
                                    fill_color = stn_reports['hly_dly_mly']['color']))
    elif hourly and daily :
        markers.append(CircleMarker(location = (lat, lon), radius = 1,
                                    color = stn_reports['hly_dly']['color'], 
                                    fill_color = stn_reports['hly_dly']['color']))
    elif hourly and monthly:
        markers.append(CircleMarker(location = (lat, lon), radius = 1,
                                    color = stn_reports['hly_mly']['color'], 
                                    fill_color = stn_reports['hly_mly']['color']))
    elif hourly :
        markers.append(CircleMarker(location = (lat, lon), radius = 1,
                                    color = stn_reports['hly']['color'], 
                                    fill_color = stn_reports['hly']['color']))
    elif daily and monthly:
        markers.append(CircleMarker(location = (lat, lon), radius = 1,
                                    color = stn_reports['dly_mly']['color'],
                                    fill_color = stn_reports['dly_mly']['color']))
    elif monthly:
        markers.append(CircleMarker(location = (lat, lon), radius = 1,
                                    color = stn_reports['mly']['color'],
                                    fill_color = stn_reports['mly']['color']))
    else:
        markers.append(CircleMarker(location = (lat, lon), radius = 1, 
                                    color = stn_reports['other']['color'], 
                                    fill_color = stn_reports['other']['color']))

    name = stnID_df[stnID_df['Climate ID'] == stn]['Name'].values[0]
    message.append(f'STN Info :\n\tname : {name}\n\tClimate ID : \n\tTC ID : \n\t')

        
# lat, lon = (62.20, -75.63)
# print(lat, lon)
# icon = AwesomeIcon(name="fa fa-circle", marker_color="gray", icon_color="darkgray")
# marker = Marker(icon=icon, location=(lat, lon))

# circle_marker = CircleMarker(location = (lat, lon), radius = 1, color = 'red', fill_color = 'red')
# circle_marker.weight = 6
# circle_marker.fillOpacity = 0
````

````python
center = [53, -70]
zoom = 4

mapnik = basemap_to_tiles(basemaps.OpenStreetMap.Mapnik)
mapnik.base = True
mapnik.name = 'Mapnik Layer'

# toner = basemap_to_tiles(basemaps.Stamen.Toner)
# toner.base = True
# toner.name = 'Toner Layer'

bzh = basemap_to_tiles(basemaps.OpenStreetMap.BZH)
bzh.base = True
bzh.name = 'BZH layer'

grey =  basemap_to_tiles(basemaps.CartoDB.Positron)
grey.base = True
grey.name = 'Grey layer'

topo =  basemap_to_tiles(basemaps.Gaode.Satellite)
topo.base = True
topo.name = 'Topo layer'

strava =  basemap_to_tiles(basemaps.Strava.Ride)
strava.base = True
strava.name = 'Gaode layer'

m = Map(center=center, zoom=zoom, layers=[mapnik, bzh, topo, strava, grey],
        layout=Layout(width='50%', height='500px'))

m.add_control(LayersControl())
````

````python
# marker = Marker(location=(53, -70))
# m += marker

# message1 = HTML()
# message2 = HTML()
# message1.value = "Try clicking the marker!"
# message2.value = "Hello <b>World</b>"
# message2.placeholder = "Some HTML"
# message2.description = "Some HTML"

# # Popup with a given location on the map:
# popup = Popup(
#     location=center,
#     child=message1,
#     close_button=False,
#     auto_close=False,
#     close_on_escape_key=False
# )
# m.add(popup)

# # Popup associated to a layer
# marker.popup = message2

for index, marker in enumerate(markers):
    m += marker

    message2 = HTML()
    message2.value = message[index]
    marker.popup  = message2
````

````python
legend = LegendControl({})

# Set/Get legend title
# legend.title = "Stations Reports"  # Set title
# legend.title  # Get title

# # Set/Get legend content
# legend.legend = {}  # Set content
# legend.legend  # Get content

for report in stn_reports :
    legend.add_legend_element(stn_reports[report]['name'],stn_reports[report]['color'])  # Add a legend element
   
# legend position
legend.position = "topright"  # Set position
legend.position  # Get current position

m.add(legend)
````
